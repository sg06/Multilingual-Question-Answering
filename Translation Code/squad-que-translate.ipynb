```python
import datasets
from datasets import load_dataset, concatenate_datasets
```

```python
squad = load_dataset("squad")
print(squad)
```

```python
dataset1 = load_dataset("ai4bharat/IndicQuestionGeneration", "bn")
print(dataset1)
```

```python
d_tel = load_dataset("ai4bharat/IndicQuestionGeneration", "te")
print(d_tel)
```

```python
from datasets import concatenate_datasets
dataset1['train']=concatenate_datasets([dataset1['train'],d_tel['train']])
dataset1['validation']=concatenate_datasets([dataset1['validation'],d_tel['validation']])
dataset1['test']=concatenate_datasets([dataset1['test'],d_tel['test']])
```

```python
ids={}
ids['train']={}
ids['validation']={}

for i in range(len(squad['train'])):
    ids['train'][squad['train'][i]['id']]=i

for i in range(len(squad['validation'])):
    ids['validation'][squad['validation'][i]['id']]=i
    
```

```python
import copy
```

```python
def create_dataset(ds):
    dict_lang={}
    dict_lang['data']=[]
    for i in range(len(ds)):
        idx=-1
        type=None
        if ds[i]['squad_id'] in ids['train']:
            idx=ids['train'][ds[i]['squad_id']]
            type='train'
        if ds[i]['squad_id'] in ids['validation']:
            idx=ids['validation'][ds[i]['squad_id']]
            type='validation'
        if idx==-1:
            continue
        dict_temp={}
        dict_temp['id']=ds[i]['squad_id']+'-'+ds[i]['id']
        dict_temp['title']=squad[type][idx]['title']
        dict_temp['context']=squad[type][idx]['context']
        dict_temp['question']=ds[i]['question']
        dict_temp['answers']=copy.deepcopy(squad[type][idx]['answers'])
        dict_lang['data'].append(dict_temp)
        
    print(print('dataset size',len(dict_lang['data'])))
    return dict_lang
        
```

```python
import json
with open("squad_be_train1.json", "w") as outfile:
    json.dump(create_dataset(dataset1['train']), outfile)
with open("squad_be_val1.json", "w") as outfile:
    json.dump(create_dataset(dataset1['validation']), outfile)
with open("squad_be_test1.json", "w") as outfile:
    json.dump(create_dataset(dataset1['test']), outfile)
```
