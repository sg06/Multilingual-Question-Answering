```python
import datasets
from datasets import load_dataset, concatenate_datasets
```

```python
squad = load_dataset("squad")
print(squad)
```

```python
dataset1 = load_dataset("ai4bharat/IndicQuestionGeneration", "bn")
print(dataset1)
```

```python
d_tel = load_dataset("ai4bharat/IndicQuestionGeneration", "te")
print(d_tel)
```

```python
from datasets import concatenate_datasets
dataset1['train']=concatenate_datasets([dataset1['train'],d_tel['train']])
dataset1['validation']=concatenate_datasets([dataset1['validation'],d_tel['validation']])
dataset1['test']=concatenate_datasets([dataset1['test'],d_tel['test']])
```

```python
def create_dataset(ds):
    dict_lang={}
    dict_lang['data']=[]
    for i in range(len(ds)):
        s=ds[i]['context'].replace('-',' ')
        t=ds[i]['answer'].replace('-',' ')
        idx=s.find(t)
        if idx==-1:
            continue
        dict_temp={}
        dict_temp['id']=ds[i]['squad_id']+'-'+ds[i]['id']
        dict_temp['title']='translated_text'
        dict_temp['context']=s
        dict_temp['question']=ds[i]['question']
        dict_temp['answers']={'text':[t],'answer_start':[idx]}
        dict_lang['data'].append(dict_temp)
        
    print(print('dataset size',len(dict_lang['data'])))
    return dict_lang
        
```

```python
import json
with open("squad_be_train.json", "w") as outfile:
    json.dump(create_dataset(dataset1['train']), outfile)
with open("squad_be_val.json", "w") as outfile:
    json.dump(create_dataset(dataset1['validation']), outfile)
with open("squad_be_test.json", "w") as outfile:
    json.dump(create_dataset(dataset1['test']), outfile)
```

```python
dataset1['train']=load_dataset('json',data_files='squad_be_train.json',field='data',features=squad['train'].features)['train']
dataset1['validation']=load_dataset('json',data_files='squad_be_val.json',field='data',features=squad['train'].features)['train']
dataset1['test']=load_dataset('json',data_files='squad_be_test.json',field='data',features=squad['train'].features)['train']
```
